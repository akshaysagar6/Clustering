```python
import pandas as pd
```

```python
iris = pd.read_csv('../datasets/iris.csv')
```

```python
iris.head()
```

```python
df = iris.drop('species', axis=1)
```

```python
df.head()
```

### Build the model

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(min_samples=12, eps=0.5 )
```

```python
model.fit(df)
```

```python
model.labels_
```

```python
iris['Label'] = model.labels_
```

```python
iris._get_numeric_data
```

```python
iris
```

```python
grp = iris.groupby('Label')
```

```python
for x,y in grp:
    print(y)
    print('_'*70)
```

```python
grp.agg('mean')
```

```python
sgrp = iris.groupby('species')
```

```python
sgrp.agg('mean')
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(df, iris['species'])
```

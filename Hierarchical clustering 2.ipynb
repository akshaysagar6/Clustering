### Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
stdan=StandardScaler()
new_df=pd.DataFrame(stdan.fit_transform(df),columns=df.columns)
new_df
```

### Normalization

```python
def normal(i):
    x=(i-i.min())/(i.max()-i.min())
    return x
```

```python
new_df2=df.apply(normal)
```

```python
new_df2
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(new_df2,method='average',metric='euclidean')
plt.figure(figsize=(10,5))
sch.dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agglo=AgglomerativeClustering(n_clusters=5,linkage='average',affinity="euclidean").fit(new_df2)
```

```python
cluster_labels=pd.Series(agglo.labels_)
```

```python
cluster_labels
```

```python
df1['cluster']=cluster_labels
```

```python
df1
```

```python
pd.crosstab(df1['Univ'],df1['cluster'])
```

```python
grp=df.groupby('cluster')
for x, y in grp:
    print(x)
    print('-'*30)
    print(y)
    print('-'*30)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x=np.random.uniform(0,1,1000)
y=np.random.uniform(0,1,1000)
```

```python
df=pd.DataFrame(columns=[['X','Y']])
df['X']=x
df['Y']=y
```

```python
df
```

```python
plt.scatter(df['X'],df['Y'])
```

```python
model=KMeans(n_clusters=2).fit(df)
```

```python
labels=model.labels_
labels
```

```python
df['label1']=labels
```

```python
model.inertia_
```

```python
df.plot(x='X',y='Y',c=labels,kind='scatter',s=50,cmap=plt.cm.coolwarm_r)
```

```python
df1=pd.read_csv(r"C:\Users\aksha\Downloads\Universities (1).csv")
df1
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
norm=MinMaxScaler()
df1.drop('Univ',axis=1,inplace=True)
df2=pd.DataFrame(norm.fit_transform(df1),columns=df1.columns)
```

```python
df2
```

```python
wcss=[]
for i in range(1,10):
    model=KMeans(n_clusters=i).fit(df2)
    wcss.append(model.inertia_)
    
```

```python
plt.plot(np.arange(1,10),wcss,marker='*',linestyle='dotted')
plt.ylabel('WCSS')
plt.xlabel('Cluster')
plt.show()
```
